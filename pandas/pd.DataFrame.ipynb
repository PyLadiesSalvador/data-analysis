# pd.DataFrame
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

```python
import pandas as pd
```

## Criar DataFrame a partir de lista

```python
lista = [1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]
```

```python
pd.DataFrame(lista)
```

```python
# colocar nome na coluna e definir tipo de dados
pd.DataFrame(lista, columns = ['despesa'], dtype = "float")
```

```python
# usando lista de listas para passar os dados de cada linha
pd.DataFrame([
    ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'],
    [1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]
])
```

```python
pd.DataFrame([[1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]],

columns = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'])
```

## Criar DataFrame a partir de dicionário

```python
gasto_mensal = {"mês" : ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'],
                "despesas" :[1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]}
 
    
pd.DataFrame(gasto_mensal)
```

```python
# definir os valores do índice   
pd.DataFrame({"despesas_2020" :[1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]},
            index = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'])
```

```python
df = pd.DataFrame({"despesas_2020" :[1000, 2000, 1500, 1000, 1200, 800, 1100, 1800, 1300, 1000,2100, 1500]},
            index = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'])
df
```
