## Filltrar linhas de acordo com condições de colunas tipo string

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'nome':['Joana Silva', 'andre santos', None, 'Leonardo Oliveira', 'Ricardo Souza', 'Camila Alves',
           'Suzane Borges', 'João Dias', 'Roberto Silva'],
    'cpf':['000.000.000-00', '000.000.005-00', '999.999.992-00', '999.999.993-00',
          '999.999.004-00', '000.888.995-99',  '000.888.991-99',  '000.888.994-99',
           '000.888.992-99'],
    'sexo':['feminino', 'masculino', 'feminino', 'masculino', 'masculino', 'feminino', 'feminino', 'masculino', 'masculino'],
    'nota':[9.5, 5.8, 4.7, 6.5, 7, 6.5, 8.3, 9.2, 5.5]
})
df
```

```python
df['sexo'] == 'feminino'
```

```python
df[df['sexo'] == 'feminino']
```

### igual, diferente, antes, depois (ordenado alfabeticamente)

```python
# valor da coluna igual a valor selecionado

df[df['sexo'] == 'feminino']
```

```python
# valor da coluna diferente do valor selecionado

df[df['sexo'] != 'feminino']
```

```python
df[df['sexo'] < 'feminino']
```

```python
df[df['nome'] > 'Joana Silva']
```

```python
df.sort_values(by =  'nome')
```

```python
df[df['nome'] <= 'Joana Silva']
```

```python
df[df['nome'] > 'Joana']
```

### Entre

https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html

```python
df[df['nome'].between('Camila Alves', 'Ricardo Souza')]
```

```python
df[df['nome'].between('Camila Alves', 'Ricardo Souza', inclusive = 'left')]
```

### Retorna valores listados

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html

```python
df[df['nome'].isin(['Joana Silva', 'Camila Alves'])]
```

### Começa com, Termina com

https://pandas.pydata.org/docs/reference/api/pandas.Series.str.startswith.html

https://pandas.pydata.org/docs/reference/api/pandas.Series.str.endswith.html

```python
df[df['nome'].str.startswith(pat = 'L', na = True)]
```

```python
df[df['nome'].str.endswith(pat = 'a', na = False)]
```

```python
# multiplas opções para busca

df[df['nome'].str.startswith(pat = ('L', 'R'), na = False)]
```

```python
df[df['nome'].str.startswith(pat = ('L', 'Ro'), na = False)]
```

### Contém

https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html

```python
df[df['nome'].str.contains('L', na = False)]
```

```python
df[df['nome'].str.contains('D', na = False)]
```

```python
df[df['nome'].str.contains('D', case = False, na = False)]
```

### Nulos, não nulos

```python
df[df['nome'].isna()]
```

```python
df[df['nome'].notna()]
```

### Negar a condição

```python
df[~df['nome'].str.contains('L', na = False)]
```

```python
df[~df['nome'].isin(['Joana Silva', 'Camila Alves'])]
```

### Filtrar com mais de uma condição

```python
# o e tem que ser o caracter &
df[(df['sexo'] == 'feminino') & (df['cpf'].str[-4] != '4')]
```

```python
# o ou tem que ser o caracter |
df[(df['sexo'] == 'feminino') | (df['cpf'].str[-4] != '4')]
```

```python
cond1 = df['sexo'] == 'feminino'
cond2 = df['cpf'].str[-4] != '4'

df[cond1 & cond2]
```
