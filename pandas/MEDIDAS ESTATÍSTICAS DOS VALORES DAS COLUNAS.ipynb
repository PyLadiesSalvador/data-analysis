## MEDIDAS ESTATÍSTICAS DOS VALORES DAS COLUNAS

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html

```python
import pandas as pd
```

```python
df = pd.read_csv("pyladies.csv",
            encoding = "latin", sep = ";",
            decimal = ",", 
            skipinitialspace = True, 
            na_values = { 'idade':'-'}, 
            parse_dates = ['data atualização'], 
            dayfirst = True, 
            dtype = {"meses tratamento":'int8', "sexo":'category', 'nome':str})
df
```

```python
df.info()
```

```python
df.describe() # POR PADRÃO MOSTRA APENAS AS COLUNAS COM VALORES NUMÉRICOS pencentis do boxplot
```

```python
df.describe(percentiles=[.10,.30,.40]) # PODEMOS DEFINIR OS PERCENTIS QUE QUEREMOS
```

```python
df.describe(include = 'all', datetime_is_numeric=True) # PODEMOS INCLUIR TODAS AS COLUNAS, MESMO AS NÃO NUMÉRICAS
```

```python
df.describe(include = ['object', 'category']) # PODEMOS ESCOLHER OS TIPOS DE DADOS QUE QUEREMOS 
```

```python
df.describe(exclude = [ 'category'], datetime_is_numeric=True) # PODEMOS ESCOLHER OS TIPOS DE DADOS QUE NÃO QUEREMOS 
```
