CAMPOS NUMÉRICOS 

Problemas encontrados para que o pandas reconheça os campos numéricos:

1 - o separador de decimal utilizado é a vírgula, que é diferente do padrão de float do python que é o ponto.

2 - espaços e traços em células onde não há informação dos valores.


```python
df = pd.read_csv(r".\Users\pyladies.csv", 
            encoding = 'latin', 
            sep = ";",            
            decimal = ",", # indica o caractere seperador de decimal no arquivo
            #skipinitialspace = True, # apaga os espaços antes dos caracteres do campo
            na_values = ['-'," "]) # indica quais caracteres devem ser considerados como nulos
df
```

```python
df.dtypes
```

```python
df = pd.read_csv(r".\Users\pyladies_sem_nomes_colunas.csv", 
            encoding = 'latin', 
            sep = ";",     
            header = None, # informa que a tabela não tem os nomes das colunas
            names= ['nome', 'documento', 'sexo', 'data atualização', 'altura', 'peso',
       'idade', 'meses tratamento', 'nutricionista'], # indica nomes para as colunas
            decimal = ",", # indica o caractere seperador de decimal no arquivo
            skipinitialspace = True, # apaga os espaços antes dos caracteres do campo
            na_values = { 'idade':'-'}, # indica quais caracteres em quais colunas devem ser considerados como nulos
            parse_dates = ['data atualização'], # indica o campo ou campos com informação de data
            dayfirst = True, # indica que a data do arquivo começa com o dia do mês
            dtype = {"meses tratamento":'int8', "sexo":'category', 'nome':str}, # define os tipos de dados que queremos em campos específicos
            usecols = ['nome', 'documento', 'sexo', 'data atualização', 'altura', 'peso',
       'idade', 'meses tratamento'] # define as colunas da tabela que queremos carregar
           )
df
```
